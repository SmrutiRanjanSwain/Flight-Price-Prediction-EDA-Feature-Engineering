```python
#importing basics libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_excel('flight_price.xlsx')
df.head()
```

```python
df.tail()
```

```python
## get the basics info about data
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
## Feature Engineering
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.info()
```

```python
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
df.head()
```

```python
## Drop Date Of Journey

df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
df['Arrival_hour']=df['Arrival_Time'].str.split(':').str[0]
df['Arrival_min']=df['Arrival_Time'].str.split(':').str[1]
```

```python
df.head(2)
```

```python
df['Arrival_hour']=df['Arrival_hour'].astype(int)
df['Arrival_min']=df['Arrival_min'].astype(int)
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.head(2)
```

```python
df['Departure_hour']=df['Dep_Time'].str.split(':').str[0]
df['Departure_min']=df['Dep_Time'].str.split(':').str[1]
```

```python
df['Departure_hour']=df['Departure_hour'].astype(int)
df['Departure_min']=df['Departure_min'].astype(int)
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df.head(2)
```

```python
df.info()
```

```python
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,np.nan:1})
```

```python
# check the null
df[df['Total_Stops'].isnull()]
```

```python
df.head(2)
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="Total_Stops", y="Price", data=df)
plt.title("Flight Price vs Total Stops")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x="Month", y="Price", data=df)
plt.title("Flight Price by Month (Seasonality)")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x="Airline", y="Price", data=df)
plt.xticks(rotation=90)
plt.title("Flight Price Distribution by Airline")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x="Airline", y="Price", data=df, estimator=np.mean, ci=None)
plt.xticks(rotation=90)
plt.title("Average Flight Price by Airline")
plt.ylabel("Average Price")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x="Month", y="Price", data=df, estimator=np.mean, ci=None, marker="o")
plt.title("Average Flight Price Trend by Month")
plt.ylabel("Average Price")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x="Source", y="Price", data=df, estimator=np.mean, ci=None, palette="muted")
plt.title("Average Flight Price by Source City")
plt.ylabel("Average Price")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x="Destination", y="Price", data=df, estimator=np.mean, ci=None, palette="muted")
plt.title("Average Flight Price by Destination City")
plt.ylabel("Average Price")
plt.show()
```

```python
plt.figure(figsize=(12,6))
top_routes = df.groupby(["Source","Destination"])["Price"].mean().sort_values(ascending=False).head(5)
top_routes.plot(kind="bar", color="coral")
plt.title("Top 5 Most Expensive Routes")
plt.ylabel("Average Price")
plt.show()
```

```python
plt.figure(figsize=(12,6))
route_counts = df.groupby(["Source","Destination"]).size().sort_values(ascending=False).head(5)
route_counts.plot(kind="bar", color="skyblue")
plt.title("Top 5 Most Frequent Routes")
plt.ylabel("Number of Flights")
plt.show()
```
